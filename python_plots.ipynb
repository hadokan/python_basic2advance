```python
import numpy as np
import matplotlib.pyplot as plt

stores = np.array(["istanbul", "ankara", "bursa",  "izmir", "eskisehir", "antalya"])
sales = np.array([1000, 800, 650, 900, 400, 350 ])

plt.plot(stores, sales)
plt.show()
```

```python
plt.scatter(stores, sales)
plt.show()
```

```python
# set colors and sizes
colors = np.array([5, 1, 2, 3, 4, 5])
sizes = np.array([100, 50, 50, 200, 15, 120])

plt.scatter(stores, sales, c=colors, s=sizes)
plt.show()
```

```python
# create a bar graph
plt.bar(stores, sales)

plt.title('Bar Graph')

plt.show()
```
