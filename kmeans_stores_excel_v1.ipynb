```python
import pandas as pd
import numpy as np
#import pyodbc
from sklearn.cluster import KMeans
from k_means_constrained import KMeansConstrained
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_excel(r'/content/gdrive/My Drive/store_data_v1.xlsx')


```

```python
df
```

```python
kmeans = KMeans(n_clusters=12)
df["CLUSTER"] = kmeans.fit(df[['SALE_CNT', "DOKMEURUN_CNT", "DOKMEURUN_QTY"]]).labels_
```

```python
df
```

```python
cluster_norm_index = df[["SALE_CNT", "DOKMEURUN_CNT", "DOKMEURUN_QTY", "CLUSTER"]].groupby("CLUSTER").mean().sort_values("SALE_CNT").index
```

```python
norm_new = {}
for i in range(len(cluster_norm_index)):
    norm_new[cluster_norm_index[i]] = i+1
df["NORM"] = df["CLUSTER"].map(norm_new)
df["NORM_NN"] = df["NORM"].map(lambda x: 3 + (x-1)*0.5)

df
df.to_excel("magaza_norm_v6.xlsx")
```
